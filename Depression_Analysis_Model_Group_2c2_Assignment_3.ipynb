```python
# Import all that is needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import seaborn as sn

```

***Reading the Data, setting up seaborn as default***


```python
# set seaborn to default graphics
sn.set()
# To read document for the course
# training data
training_data=pd.read_csv("AI-Lab-master/train.csv")
# testing data
testing_data=pd.read_csv("AI-Lab-master/test.csv")
training_data
```

***Data Cleaning***

```python
# Changing household members to unkown which can be used as a pointer
print("hi")
values = {'hh_totalmembers': "unknown","med_u5_deaths":0,"ed_expenses":0,"ed_expenses_perkid":0,"ed_schoolattend":0,
         "ed_work_act_pc":0,"ed_sch_missedpc":0,"med_u5_deaths":0}
training_data.fillna(value=values,inplace=True)
training_data.fillna("unknown",inplace=True)
training_data
```

***PREDICTIVE MODEL***

<h5>LINEAR REGRESSION</h5></br><p>Had two options to use which were Linear regression and MLP. We decided to go with linear regression as the data is simple and MLP would have worked as well but it would need us to define other factors to define for level 2 layers such as Financial stability, Basic Healthcare, Family situation and e.t.c, basically mlp would work but it would be overkill.</p></br><p>MLP would have provided the most accurate and consistent model but also considering the data is not as clean and interploation, median and mean cuould not be used to fill some the fields we went with the linear regression model</p>

```python
# Training the model
# Considering the sensitivity of some of the information and the lack of it would lead to 
# misinformation we will not use all the columns
# Columns used
# female res, age, married, children, edu, hhsize, cons_nondurable,durable investments, non durable investments
reg_model=LogisticRegression(solver='lbfgs', max_iter=500)
reg_model.fit(training_data[["femaleres","age","married","children","edu","hhsize",
                             "cons_nondurable","durable_investment","nondurable_investment"]],training_data.depressed)
```

```python
# Coeffetients of the linear regression model
print(reg_model.coef_)
print("\n")
# Y-intercept
print(reg_model.intercept_)
```

***Testing set***

```python
# clean training set
testing_data=testing_data.interpolate()
```

```python
# using a loop to fill up the table


y=0
for i in testing_data["surveyid"]:
        v1=testing_data["femaleres"][y]
        v2=testing_data["age"][y]
        v3=testing_data["married"][y]
        v4=testing_data["children"][y]
        v5=testing_data["edu"][y]
        v6=testing_data["hhsize"][y]
        v7=testing_data["cons_nondurable"][y]
        v8=testing_data["durable_investment"][y]
#         print(v2)
        v9=testing_data["nondurable_investment"][y]
        y+=1
        #predicting if they are depressed
        depressed=int(reg_model.predict([[(v1),(v2),(v3),(v4),(v5),(v6),(v7),(v8),v9]]))
        
# #        rounding up or down
#         if depressed>0.5:
#             depressed=1
#         else:
#             depressed=0
#         print(depressed)
        #changing the depression status according to depression
        testing_data["depressed"]=depressed
        
testing_data
```
