```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

datafile=pd.read_csv("AI-Lab-master/train.csv")
datafile
```

```python
women=(list(datafile.femaleres).count(1))
men=(list(datafile.femaleres).count(0))

print(women)
print(men)

df = pd.DataFrame({'lab':['women', 'men',], 'val':[women,men]})
ax = df.plot.bar(x='lab', y='val', rot=0)
ax.legend()

fig = plt.gcf()
fig.set_size_inches(10.5, 5)
fig.savefig('men_women.png', dpi=100)
```

```python
married=(list(datafile.married).count(1))
single=(list(datafile.married).count(0))

print(married)
print(single)

df = pd.DataFrame({'lab':['married', 'single/other relationship',], 'val':[married,single]})
ax = df.plot.bar(x='lab', y='val', rot=0)
ax.legend()

fig = plt.gcf()
fig.set_size_inches(10.5, 5)
fig.savefig('married_single.png', dpi=100)
```
